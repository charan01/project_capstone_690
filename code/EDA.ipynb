```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your dataset (assuming it's in a CSV file)
# Replace 'your_dataset.csv' with your actual dataset file name
#df = pd.read_csv('your_dataset.csv')

# Select a random flight ID
random_flight_id = np.random.choice(real['flight_id'].unique())

# Filter the dataset for the selected flight
flight_data = real[real['flight_id'] == 'AAL92_016']

# Extract altitude, distance, and groundspeed data
altitude = flight_data['altitude']
distance = flight_data['distance']
groundspeed = flight_data['groundspeed']

fig, axs = plt.subplots(3, 1, figsize=(10, 12))

# Plot altitude
axs[0].plot(altitude, color='blue')
axs[0].set_title('Altitude')
axs[0].set_xlabel('Data points')
axs[0].set_ylabel('Altitude')
#axs[0].set_xticks(np.arange(0, 201, 25))
#axs[0].set_xlim(0, 200)


# Plot distance
axs[1].plot(distance, color='green')
axs[1].set_title('Distance to Zurich Airport')
axs[1].set_xlabel('Data points')
axs[1].set_ylabel('Distance')

# Plot groundspeed
axs[2].plot(groundspeed, color='red')
axs[2].set_title('Groundspeed')
axs[2].set_xlabel('Data points')
axs[2].set_ylabel('Groundspeed')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
